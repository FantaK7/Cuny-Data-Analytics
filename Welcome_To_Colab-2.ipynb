```
#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Read dataset
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv')
```

```
#View data
df.head()
```

```
df.describe(include='all')
```

```
df.describe(include='all')
```

```
# Convert readmitted into binary outcome
def convert_readmitted(x):
    if x == '<30' or x == '>30':
        return "Yes"
    else:
        return "No"

df['Readmitted'] = df['readmitted'].apply(convert_readmitted)
```

```
# Check
print(df['Readmitted'].value_counts())
```

```
#Select variables
cols = df[['race', 'gender', 'age', 'weight', 'Readmitted']]
cols.head()
```

```
#clean data
cols = cols.replace('?', np.nan)

# Drop missing values for simplicity
cols= cols.dropna()

cols.isnull().sum()
```

```
#Relation between Readmission and Gender
sns.countplot(data=cols, x='gender', hue='Readmitted')
plt.title("Readmission by Gender")
plt.show()
```

More people were readmitted than not, for both men and women.
The pattern of readmission is similar for both genders.

```
#Readmission and Weight
plt.figure(figsize=(10,5))
sns.countplot(data=cols, x='weight', hue='Readmitted')
plt.xticks(rotation=90)
plt.title("Readmission by Weight")
plt.show()
```

Most patients fall in the weight ranges of 75-100, 50-75, and 100-125.
In all weight groups, more people were readmitted than not.
The highest readmissions happen in the 75-100 weight group.

```
#Readmission. and Age
plt.figure(figsize=(10,5))
sns.countplot(data=cols, x='age', hue='Readmitted')
plt.xticks(rotation=45)
plt.title("Readmission by Age Group")
plt.show()
```

Most readmissions happen in the age groups 60-70 and 70-80.
For almost all age groups, more people were readmitted  than not.
The 70-80 age group has the highest readmission count.
Younger age groups (0-10, 10-20, 20-30) have fewer patients overall and fewer readmissions.

```
#Readmission by Race

plt.figure(figsize=(10,5))
sns.countplot(data=cols, x='race', hue='Readmitted')
plt.xticks(rotation=45)
plt.title("Readmission by Race")
plt.show()
```

Caucasian patients make up the vast majority of both readmitted and non-readmitted groups..
African American patients have fewer overall cases but also show a higher number of readmissions than non-readmissions.
Other and Asian race groups have very small sample sizes, but the readmission and non-readmission counts are roughly similar.
